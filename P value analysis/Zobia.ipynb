```python
#this file contains p-value analysis for method level multiclass then after that class level multiclass
import pandas as pd
df = pd.read_csv("R3-2.csv")
df.shape
#df.drop(['is_method_level'],axis='columns',inplace=True)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd
p_value = pd.DataFrame(columns=['col','stat','p'])
cols = df.columns
for i in range(0, 9):
    input = df.iloc[:, i]
    column = cols[i]
    target = df['Priority Order']
    table = pd.crosstab(input, target)
    stat, p, dof, expected = chi2_contingency(table.values)
    print (column,stat,p)
    p_value.loc[i] = [column, stat, p]
print(p_value)
```
